```python
# 1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results@

import scipy.stats as stats
import numpy as np

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Interpretation
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-Score: {z_score}")
    print(f"P-Value: {p_value}")

    if p_value < alpha:
        print("Reject the Null Hypothesis (H0): The sample mean is significantly different from the population mean.")
    else:
        print("Fail to Reject the Null Hypothesis (H0): No significant difference between sample and population mean.")

# Example Data
sample_data = [52, 48, 49, 51, 50, 53, 47, 49, 51, 50]  # Sample data
pop_mean = 50  # Hypothetical population mean
pop_std = 2    # Known population standard deviation

# Perform Z-test
z_test(sample_data, pop_mean, pop_std)
```

```python
# 3 Implement a one-sample Z-test using Python to compare the sample mean with the population mean

import numpy as np
import scipy.stats as stats

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a one-sample Z-test to compare sample mean with population mean.

    Parameters:
    sample (array-like): Sample data.
    pop_mean (float): Population mean.
    pop_std (float): Population standard deviation (known).
    alpha (float): Significance level (default = 0.05).

    Returns:
    None (Prints Z-score, p-value, and decision)
    """
    # Compute sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Score: {z_score:.3f}")
    print(f"P-Value: {p_value:.5f}")

    # Hypothesis Test Interpretation
    if p_value < alpha:
        print("Reject the Null Hypothesis (H0): Sample mean is significantly different from the population mean.")
    else:
        print("Fail to Reject the Null Hypothesis (H0): No significant difference between sample and population mean.")

# Example: Simulated Sample Data
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=102, scale=15, size=30)  # Simulated sample from N(102,15)
pop_mean = 100  # Hypothetical population mean
pop_std = 15    # Known population standard deviation

# Perform One-Sample Z-Test
one_sample_z_test(sample_data, pop_mean, pop_std)
```

```python
# 4  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a two-tailed Z-test and visualize the decision region.

    Parameters:
    sample (array-like): Sample data.
    pop_mean (float): Population mean.
    pop_std (float): Population standard deviation (known).
    alpha (float): Significance level (default = 0.05).

    Returns:
    None (Prints Z-score, p-value, and displays a plot)
    """
    # Compute sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Score: {z_score:.3f}")
    print(f"P-Value: {p_value:.5f}")

    # Decision based on alpha
    critical_value = stats.norm.ppf(1 - alpha / 2)  # Two-tailed critical Z-value
    if abs(z_score) > critical_value:
        print("Reject the Null Hypothesis (H0): Sample mean is significantly different from the population mean.")
    else:
        print("Fail to Reject the Null Hypothesis (H0): No significant difference.")

    # Visualization
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)  # Standard normal distribution

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Standard Normal Distribution", color='black')

    # Shade rejection regions
    plt.fill_between(x, y, where=(x < -critical_value) | (x > critical_value), color='red', alpha=0.3, label="Rejection Region")

    # Mark Z-score
    plt.axvline(z_score, color='blue', linestyle='--', label=f"Z-score = {z_score:.3f}")

    # Labels and legend
    plt.axvline(-critical_value, color='red', linestyle='dashed', label=f"Critical Z = ±{critical_value:.3f}")
    plt.axvline(critical_value, color='red', linestyle='dashed')
    plt.title("Two-Tailed Z-Test Decision Region")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example: Simulated Sample Data
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=102, scale=15, size=30)  # Sample from N(102,15)
pop_mean = 100  # Population mean
pop_std = 15    # Known population standard deviation

# Perform Two-Tailed Z-Test and Plot
two_tailed_z_test(sample_data, pop_mean, pop_std)
```

```python
# 5 @ Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(alpha=0.05, effect_size=0.5, sample_size=30, pop_std=1):
    """
    Visualizes Type 1 and Type 2 Errors in hypothesis testing.

    Parameters:
    alpha (float): Significance level (default=0.05)
    effect_size (float): Difference between true mean and null mean in standard deviation units
    sample_size (int): Sample size
    pop_std (float): Population standard deviation

    Returns:
    None (Displays a plot)
    """
    # Critical Z-value for given alpha (two-tailed test)
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Standard normal distribution (Null Hypothesis H0)
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    # Alternative hypothesis distribution (shifted)
    true_mean_shift = effect_size * (pop_std / np.sqrt(sample_size))
    y_alt = stats.norm.pdf(x, true_mean_shift, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Null Hypothesis (H0)", color='black')
    plt.plot(x, y_alt, label="Alternative Hypothesis (H1)", color='blue')

    # Shade Type 1 Error (alpha) regions
    plt.fill_between(x, y, where=(x > z_critical) | (x < -z_critical), color='red', alpha=0.3, label="Type 1 Error (α)")

    # Shade Type 2 Error (Beta) region (failing to reject H0)
    beta_region = (x < z_critical) & (x > -z_critical)
    plt.fill_between(x, y_alt, where=beta_region, color='blue', alpha=0.3, label="Type 2 Error (β)")

    # Labels and legend
    plt.axvline(z_critical, color='red', linestyle='dashed', label=f"Critical Z = ±{z_critical:.3f}")
    plt.axvline(-z_critical, color='red', linestyle='dashed')
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example: Visualize Type 1 and Type 2 Errors
visualize_errors(alpha=0.05, effect_size=0.8, sample_size=30, pop_std=1)
```

```python
# 6 Write a Python program to perform an independent T-test and interpret the results

import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2, alpha=0.05):
    """
    Perform an independent (two-sample) T-test and interpret the results.

    Parameters:
    sample1 (array-like): Data for Group 1
    sample2 (array-like): Data for Group 2
    alpha (float): Significance level (default = 0.05)

    Returns:
    None (Prints T-test results and interpretation)
    """
    # Compute sample statistics
    mean1, mean2 = np.mean(sample1), np.mean(sample2)
    var1, var2 = np.var(sample1, ddof=1), np.var(sample2, ddof=1)
    n1, n2 = len(sample1), len(sample2)

    print(f"Group 1: Mean={mean1:.2f}, Variance={var1:.2f}, Size={n1}")
    print(f"Group 2: Mean={mean2:.2f}, Variance={var2:.2f}, Size={n2}")

    # Levene’s test for equal variances
    levene_stat, levene_p = stats.levene(sample1, sample2)
    equal_var = levene_p > alpha  # True if p > alpha (variances are equal)

    print(f"Levene’s Test: p-value = {levene_p:.5f} ({'Equal Variance Assumed' if equal_var else 'Unequal Variance Assumed'})")

    # Perform independent T-test
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=equal_var)

    # Print results
    print(f"T-Statistic: {t_stat:.3f}")
    print(f"P-Value: {p_value:.5f}")

    # Hypothesis Test Interpretation
    if p_value < alpha:
        print("Reject the Null Hypothesis (H0): The two groups have significantly different means.")
    else:
        print("Fail to Reject the Null Hypothesis (H0): No significant difference between the group means.")

# Example: Simulated Sample Data
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=100, scale=15, size=30)  # Group 1: Mean 100, Std 15
group2 = np.random.normal(loc=105, scale=15, size=30)  # Group 2: Mean 105, Std 15

# Perform Independent T-Test
independent_t_test(group1, group2)
```

```python
# 7 Perform a paired sample T-test using Python and visualize the comparison results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_t_test(before, after, alpha=0.05):
    """
    Perform a paired sample T-test and visualize the comparison.

    Parameters:
    before (array-like): Data for condition 1 (before treatment)
    after (array-like): Data for condition 2 (after treatment)
    alpha (float): Significance level (default = 0.05)

    Returns:
    None (Prints T-test results and displays a visualization)
    """
    # Compute differences
    differences = np.array(after) - np.array(before)

    # Perform paired T-test
    t_stat, p_value = stats.ttest_rel(before, after)

    # Print results
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)

    print(f"Mean Difference: {mean_diff:.2f}")
    print(f"Standard Deviation of Differences: {std_diff:.2f}")
    print(f"T-Statistic: {t_stat:.3f}")
    print(f"P-Value: {p_value:.5f}")

    # Hypothesis Test Interpretation
    if p_value < alpha:
        print("Reject the Null Hypothesis (H0): There is a significant difference.")
    else:
        print("Fail to Reject the Null Hypothesis (H0): No significant difference.")

    # Visualization: Boxplot and Pairwise Comparison
    plt.figure(figsize=(8, 5))

    # Boxplot
    plt.subplot(1, 2, 1)
    plt.boxplot([before, after], labels=["Before", "After"], patch_artist=True)
    plt.title("Boxplot of Before & After Data")

    # Pairwise Comparison Plot
    plt.subplot(1, 2, 2)
    plt.plot(before, after, 'bo', alpha=0.5)  # Scatter plot of paired values
    plt.plot([min(before), max(before)], [min(before), max(before)], 'r--', label="No Change Line")
    plt.xlabel("Before")
    plt.ylabel("After")
    plt.title("Paired Data Comparison")
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

# Example: Simulated Paired Data (Before vs After)
np.random.seed(42)
before_data = np.random.normal(loc=100, scale=10, size=30)  # Before treatment scores
after_data = before_data + np.random.normal(loc=2, scale=5, size=30)  # After treatment with slight improvement

# Perform Paired Sample T-Test & Visualize
paired_t_test(before_data, after_data)
```

```python
# 9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    data (array-like): Sample data
    confidence (float): Confidence level (default = 95%)

    Returns:
    tuple: (lower bound, upper bound) of confidence interval
    """
    # Sample statistics
    n = len(data)
    mean_sample = np.mean(data)
    std_sample = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)

    # Choose Z-score (large sample) or T-score (small sample)
    if n >= 30:
        critical_value = stats.norm.ppf((1 + confidence) / 2)  # Z-score
    else:
        critical_value = stats.t.ppf((1 + confidence) / 2, df=n-1)  # T-score

    # Compute margin of error
    margin_of_error = critical_value * (std_sample / np.sqrt(n))

    # Confidence interval bounds
    lower_bound = mean_sample - margin_of_error
    upper_bound = mean_sample + margin_of_error

    return lower_bound, upper_bound

# Example: Simulated Sample Data
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=25)  # Small sample (n=25)

# Compute Confidence Interval
ci_lower, ci_upper = confidence_interval(sample_data)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# 10 Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
import scipy.stats as stats

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level using sample data.

    Parameters:
    data (array-like): Sample data
    confidence (float): Confidence level (default = 95%)

    Returns:
    float: Margin of error
    """
    n = len(data)
    std_sample = np.std(data, ddof=1)  # Sample standard deviation (unbiased)

    # Choose Z-score for large samples, T-score for small samples
    if n >= 30:
        critical_value = stats.norm.ppf((1 + confidence) / 2)  # Z-score
    else:
        critical_value = stats.t.ppf((1 + confidence) / 2, df=n-1)  # T-score

    # Compute Margin of Error
    margin_error = critical_value * (std_sample / np.sqrt(n))
    return margin_error

# Example: Simulated Sample Data
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=25)  # Small sample (n=25)

# Compute Margin of Error
moe = margin_of_error(sample_data)
print(f"Margin of Error (95% Confidence): {moe:.2f}")
```

```python
# 11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, likelihood, false_positive):
    """
    Applies Bayes' Theorem to calculate the posterior probability.

    Parameters:
    prior (float): Prior probability of hypothesis P(H)
    likelihood (float): Probability of evidence given hypothesis P(E|H)
    false_positive (float): Probability of false positives P(E|~H)

    Returns:
    float: Posterior probability P(H|E)
    """
    not_prior = 1 - prior  # P(~H), probability of NOT having the disease
    marginal_likelihood = (likelihood * prior) + (false_positive * not_prior)  # P(E)

    posterior = (likelihood * prior) / marginal_likelihood
    return posterior

# Given probabilities
prior_prob = 0.01  # 1% of people have the disease
likelihood = 0.95  # Test detects the disease correctly 95% of the time
false_positive_rate = 0.05  # Test gives a false positive 5% of the time

# Compute Bayesian probability
posterior_prob = bayes_theorem(prior_prob, likelihood, false_positive_rate)

print(f"Probability of having the disease given a positive test: {posterior_prob:.4f}")
```

```python
# 13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

import numpy as np
import pandas as pd
import scipy.stats as stats

def compute_expected_frequencies(observed):
    """
    Compute expected frequencies for a given observed contingency table.

    Parameters:
    observed (numpy array or list of lists): Contingency table of observed values

    Returns:
    numpy array: Expected frequencies table
    """
    observed = np.array(observed)  # Convert to NumPy array
    row_totals = observed.sum(axis=1)  # Sum of rows
    col_totals = observed.sum(axis=0)  # Sum of columns
    grand_total = observed.sum()  # Total sum

    # Compute expected frequencies
    expected = np.outer(row_totals, col_totals) / grand_total
    return expected

# Sample Observed Data (Contingency Table)
observed_data = np.array([[30, 10],  # Males (Like, Dislike)
                          [20, 40]]) # Females (Like, Dislike)

# Compute Expected Frequencies
expected_frequencies = compute_expected_frequencies(observed_data)

# Convert to DataFrame for better visualization
df_expected = pd.DataFrame(expected_frequencies, columns=['Like', 'Dislike'], index=['Male', 'Female'])

# Print results
print("Observed Frequencies:")
print(pd.DataFrame(observed_data, columns=['Like', 'Dislike'], index=['Male', 'Female']))
print("\nExpected Frequencies:")
print(df_expected)
```

```python
# 16 Implement an F-test using Python to compare the variances of two random samples

import numpy as np
import scipy.stats as stats

def perform_f_test(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two independent samples.

    Parameters:
    sample1 (array-like): First sample data
    sample2 (array-like): Second sample data
    alpha (float): Significance level (default: 0.05)

    Returns:
    None (Prints the results)
    """
    # Calculate sample variances
    var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased)
    var2 = np.var(sample2, ddof=1)  # Sample variance (unbiased)

    # Ensure F-statistic is always >1 by dividing larger variance by smaller variance
    if var1 > var2:
        F_stat = var1 / var2
        dfn, dfd = len(sample1) - 1, len(sample2) - 1
    else:
        F_stat = var2 / var1
        dfn, dfd = len(sample2) - 1, len(sample1) - 1

    # Compute the p-value
    p_value = 1 - stats.f.cdf(F_stat, dfn, dfd)

    # Print results
    print(f"F-statistic: {F_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print("\nReject the Null Hypothesis: The variances are significantly different.")
    else:
        print("\nFail to Reject the Null Hypothesis: The variances are not significantly different.")

# Simulate two random samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std=15

# Perform the F-Test
perform_f_test(sample1, sample2)
```

```python
# 17 Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results

import numpy as np
import scipy.stats as stats

def perform_anova(*groups):
    """
    Performs a one-way ANOVA test to compare means of multiple independent groups.

    Parameters:
    groups (list of arrays): Multiple groups of numerical data

    Returns:
    None (Prints the results)
    """
    # Perform ANOVA test
    F_stat, p_value = stats.f_oneway(*groups)

    # Print results
    print(f"F-statistic: {F_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    alpha = 0.05  # Significance level
    if p_value < alpha:
        print("\nReject the Null Hypothesis: At least one group mean is significantly different.")
    else:
        print("\nFail to Reject the Null Hypothesis: No significant difference between group means.")

# Simulate three random groups
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10

# Perform ANOVA
perform_anova(group1, group2, group3)
```

```python
# 19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import scipy.stats as stats

def check_anova_assumptions(groups):
    """
    Checks the assumptions of ANOVA: normality (Shapiro-Wilk test) and equal variance (Levene's test).

    Parameters:
    groups (list of arrays): Multiple groups of numerical data

    Returns:
    None (Prints results of tests)
    """
    alpha = 0.05  # Significance level

    # 🔹 1. Normality Test (Shapiro-Wilk)
    print("🔹 Checking Normality (Shapiro-Wilk Test):")
    for i, group in enumerate(groups, start=1):
        stat, p_value = stats.shapiro(group)
        print(f"   Group {i}: W={stat:.4f}, p={p_value:.4f}")
        if p_value < alpha:
            print(f"   ❌ Group {i} is NOT normally distributed (Reject H0).")
        else:
            print(f"   ✅ Group {i} is normally distributed (Fail to Reject H0).")

    # 🔹 2. Equal Variance Test (Levene's Test)
    print("\n🔹 Checking Equal Variance (Levene's Test):")
    stat, p_value = stats.levene(*groups)
    print(f"   Levene's Test: W={stat:.4f}, p={p_value:.4f}")
    if p_value < alpha:
        print("   ❌ Variances are NOT equal (Reject H0).")
    else:
        print("   ✅ Variances are equal (Fail to Reject H0).")

    print("\n🔹 3. Independence is assumed based on study design.")

# 🔹 Simulating three random groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10

# 🔹 Check ANOVA assumptions
check_anova_assumptions([group1, group2, group3])
```

```python
# 24  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
import scipy.stats as stats

# 🔹 Step 1: Simulate Sample Data (Normally Distributed)
np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, Std Dev=5, Sample Size=30

# 🔹 Step 2: Define Hypothesized Population Variance
sigma_0_squared = 5**2  # Hypothesized population variance (std dev = 5)

# 🔹 Step 3: Compute Sample Variance
sample_variance = np.var(sample, ddof=1)  # Sample variance with Bessel's correction
n = len(sample)  # Sample size

# 🔹 Step 4: Compute Chi-Square Test Statistic
chi_square_stat = (n - 1) * sample_variance / sigma_0_squared

# 🔹 Step 5: Compute p-value (Two-tailed test)
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df=n-1), 1 - stats.chi2.cdf(chi_square_stat, df=n-1))

# Print Results
print(f"🔹 Sample Variance = {sample_variance:.4f}")
print(f"🔹 Chi-Square Statistic = {chi_square_stat:.4f}")
print(f"🔹 p-value = {p_value:.4f}")

# 🔹 Step 6: Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("✅ Reject the null hypothesis: The population variance is significantly different from the hypothesized variance.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference in population variance.")
```

```python
# 25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
import scipy.stats as stats

# 🔹 Step 1: Define Sample Data (Success Counts & Sample Sizes)
x1, n1 = 120, 300  # Group 1: 120 successes out of 300
x2, n2 = 100, 280  # Group 2: 100 successes out of 280

# 🔹 Step 2: Compute Sample Proportions
p1 = x1 / n1
p2 = x2 / n2

# 🔹 Step 3: Compute Pooled Proportion
p_hat = (x1 + x2) / (n1 + n2)

# 🔹 Step 4: Compute Z-Score
z_stat = (p1 - p2) / np.sqrt(p_hat * (1 - p_hat) * (1/n1 + 1/n2))

# 🔹 Step 5: Compute P-Value (Two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Print Results
print(f"🔹 Sample Proportions: p1 = {p1:.4f}, p2 = {p2:.4f}")
print(f"🔹 Z-Statistic = {z_stat:.4f}")
print(f"🔹 p-value = {p_value:.4f}")

# 🔹 Step 6: Interpretation
alpha = 0.05
if p_value < alpha:
    print("✅ Reject the null hypothesis: The proportions are significantly different.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference in proportions.")
```

```python
# 26 Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# 🔹 Step 1: Generate Two Random Datasets (Normal Distribution)
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std Dev=10, Size=30
group2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std Dev=15, Size=30

# 🔹 Step 2: Compute Sample Variances
var1 = np.var(group1, ddof=1)  # Sample variance of group 1
var2 = np.var(group2, ddof=1)  # Sample variance of group 2

# 🔹 Step 3: Compute F-Statistic
F_statistic = var1 / var2

# 🔹 Step 4: Compute p-value (Two-tailed test)
df1, df2 = len(group1) - 1, len(group2) - 1
p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))

# Print Results
print(f"🔹 Sample Variances: var1 = {var1:.4f}, var2 = {var2:.4f}")
print(f"🔹 F-Statistic = {F_statistic:.4f}")
print(f"🔹 p-value = {p_value:.4f}")

# 🔹 Step 5: Interpretation
alpha = 0.05
if p_value < alpha:
    print("✅ Reject the null hypothesis: The variances are significantly different.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference in variances.")

# 🔹 Step 6: Visualization of Distributions
plt.figure(figsize=(8, 5))
sns.histplot(group1, color='blue', label="Group 1 (σ=10)", kde=True, bins=10, alpha=0.6)
sns.histplot(group2, color='red', label="Group 2 (σ=15)", kde=True, bins=10, alpha=0.6)
plt.axvline(np.mean(group1), color='blue', linestyle='dashed', label="Mean Group 1")
plt.axvline(np.mean(group2), color='red', linestyle='dashed', label="Mean Group 2")
plt.legend()
plt.title("Distribution of Two Groups with Different Variances")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# 27 Perform a Chi-square test for goodness of fit with simulated data and analyze the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 🔹 Step 1: Simulate Observed Data (e.g., Survey Responses)
observed = np.array([50, 30, 20])  # Counts of three categories

# 🔹 Step 2: Define Expected Distribution (Assuming Equal Probability)
total = sum(observed)
expected = np.array([total / len(observed)] * len(observed))  # Uniform distribution

# 🔹 Step 3: Perform Chi-Square Goodness-of-Fit Test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# 🔹 Step 4: Print Results
print(f"🔹 Observed Frequencies: {observed}")
print(f"🔹 Expected Frequencies: {expected}")
print(f"🔹 Chi-Square Statistic = {chi2_stat:.4f}")
print(f"🔹 p-value = {p_value:.4f}")

# 🔹 Step 5: Interpretation
alpha = 0.05
if p_value < alpha:
    print("✅ Reject the null hypothesis: The observed distribution is significantly different from expected.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference between observed and expected distributions.")

# 🔹 Step 6: Visualization
categories = ['Category 1', 'Category 2', 'Category 3']
x = np.arange(len(categories))

plt.bar(x - 0.2, observed, width=0.4, label='Observed', color='blue', alpha=0.6)
plt.bar(x + 0.2, expected, width=0.4, label='Expected', color='red', alpha=0.6)
plt.xticks(x, categories)
plt.ylabel("Frequency")
plt.title("Observed vs Expected Frequencies")
plt.legend()
plt.show()
```
